### Task 1: Coding a Perceptron: going forward

#### Introduction
A Perceptron is an ANN with a single node. It uses $n$ input features, $n$ weights, a $bias$ term and an activation function to compute an output. The sum of the *x_i* input features multiplied by their respective *w_i* weights is then added to a $bias$ term before being passed to an activation function which gives us our final $output$:

$$output=ActivationF(bias+\sum\limits_{i=1}^{n}{x_i*w_i})$$



![Perceptron](perceptron.png)

Looking at our classification problem:
 * we have two input features: $x$ and $y$
 * we have one output feature: $Group$
 
Since our target output feature is discrete and binary (only two possible values), we want an activation function that will turn whichever continuous value is passed to it into something we can interpret as a probability of being in the first or second group. For this exercise, we will use the sigmoid function, which will provide us with a value between 0 and 1, and interpret this $output$ value as the probability that the input features belong to a data point for which $Group=2$. Since this is a binary classification problem, the probability that the input features belong to a data point for which $Group=1$ is $1-output$.

The sigmoid function is defined like this:
$$SigmoidF(x)=\frac{1}{1+e^{-x}}$$

In order to calculate $e^{-x}$, we can use the numpy *np.exp()* function.

### Task 2: Coding a Perceptron: gradient descent

#### Introduction
Like for our polynomial model in the previous E-tivity, we will use gradient descent to update our weights and bias over a number iterations and slowly improve our model.

In order to implement a gradient descent, we will need a number of things:
 * a loss function
 * a way to calculate gradients for our weights/bias
 * a weights/bias update strategy

**Loss function:** rather than looking at the final class prediction, we will be looking at the model's output after the activation function. Since we defined that $[0,1]$ interval value as the probability that the input features correspond to Group 2, we want that value to be as close to $1$ as possible for Group 2 data points, and as close to $0$ as possible for Group 1 data points. Defining individual error as:

$$output_{actual}-output_{predicted}$$

We will keep things simple by simply defining our loss as the sum of squared errors, so for a training set of $n$ samples:

$$Loss=\sum\limits_{i=1}^{n}{{(output_{i_{actual}}-output_{i_{predicted}})}^2}$$

**Gradient calculation:** rather than using calculus, we will numerically evaluate our gradients. For a given weight $w$ and a small enough $\epsilon$ value, if we adjust $w$ by plus or minus $\epsilon$ and calculate the loss in both cases, we can compute the gradient like this:

$$gradient_w=\frac{Loss_{w+\epsilon}-Loss_{w-\epsilon}}{2\epsilon}$$

It's not terribly efficient to calculate our loss that many times with a slightly different model, but in this simple case it'll do just fine and will spare us the calculation of partial derivatives. As a side thought that is **not** part of the E-tivity: can you understand why/how this calculation works?

**Weight/bias update strategy:** as in our polynomial regression model, we will define a learning rate $lr$ and update each weight/bias in our perceptron using the following formula:
$$w_{new}=w_{current}-lr\times gradient_w$$

### Tasks 1 and 2 implementation
Both tasks 1 and 2 are implemented as a class named Perceptron. This class defines all methods from both tasks, see the comments in the code.

```python
class Perceptron():
    '''
    Implements a perceptron - an ANN with one node without hidden layers.
    It receives two inputs and produces one discrete output.
    The activation function is defined as a sigmoid.
    It is defined with a set of weights equal to the number of input features
    and also has a bias term (as a part of the weights vector).
    '''

    # group assignment percentage threshold for discrete predictions
    grp_threshold = 0.5
    
    def __init__(self):
        '''
        Constructor. Does not expect any params.
        '''

        # Hyperparameters
        # 2 inputs
        self.inputLayerSize = 2
        # outputs - for future use?
        self.outputLayerSize = 1
        
        # The constructor creates a set of weights equal to the number of input features
        # and it also adds another column to that vector that represents the bias term.
        # The input features then also contain one extra column filled with "1".
        # This little trick enables easier matrix operations and gradient calculations 
        # Weights and bias are initialized to 0 here.
        # However, weights can also be changed using the function set_params()
        self.weights = np.zeros(shape=(1, self.inputLayerSize + 1))


    def sigmoid(self, z: float) -> np.ndarray | float:
        '''
        The sigmoid formula (1 / (1 + exp(-z)) for activation function
        :: Params
        z - float or numpy array of values
        :: Returns
        A sigmoid calculation as a float value or a numpy array, based on the input type
        '''
        
        # Apply sigmoid activation function to scalar, vector, or matrix
        return 1 / (1 + np.exp(-z))


    def sum_input(self, inputs: np.ndarray) -> float:
        '''
        Sums inputs and weights of the perceptron = sum(w1x1+w2x2+...+bias)
        Bias is the last column in the weights vector.
        Input features also have one extra column filled with "1" which is for the bias.
        ::Params
        inputs - numpy array of input features + extra column with "1" for the bias
        ::Returns
        A float value representing the dot product of the input features and weight(bias) vector
        '''
        
        out_sum = np.dot(inputs, self.weights)
        
        return out_sum

    
    def loss(self, probs_actual, probs_predicted) -> float:
        '''
        Implements our loss function: sum(prob_actual - prob_predicted)^2
        Probabilities are in the interval [0,1] 
        ::Params
        probs_actual - actual probabilities of our data
        probs_predicted - predicted probabilities of our model
        ::Returns
        A float value representing the loss value
        '''
    
        loss_val = np.sum(np.square(np.subtract(probs_actual, probs_predicted)))
        
        return loss_val
    

    def evaluate(self, inputs: np.ndarray) -> np.ndarray:
        '''
        Given numpy arrays of input features it calculates a final output
        (or numpy array of outputs) between 0 and 1, which represents the probability
        # that the input(s) are from Group 2.
        ::Params
        inputs - numpy array of input features
        ::Returns
        A numpy array of probabilities for input features
        '''
    
        # Sum of the inputs multiplied by weights with added bias
        sum_inputs = self.sum_input(inputs)
        
        # Sigmoid as our probability algorithm
        eval_res = self.sigmoid(sum_inputs)
        
        # Shorter version
        # eval_res = np.array([self.sigmoid(sum_i) for sum_i in self.sum_input(inputs, weights)])

        return eval_res

  
    def predict(self, inputs: np.ndarray) -> np.ndarray:
        '''
        Evaluates probability of the input features based on weights/bias
        and converts it to a group prediction:
        It returns 1 if probability<.5, returns 2 if probability>.5 and 0 if probability=.5
        ::Params
        inputs - numpy array of input features        
        '''

        probs = self.evaluate(inputs)

        predictions = np.zeros_like(probs)
        # Apply the conditions for thresholds
        predictions[probs < self.grp_threshold] = 1
        predictions[probs > self.grp_threshold] = 2

        return predictions


    def gradients(self, inputs, probs_actual, e = 1e-4) -> np.ndarray:
        '''
        Calculates numerical gradients based on the formula given in the task instructions.
        Calculates gradients not only for weights but also for bias (which is a part of the
        weight vector)
        The most difficult part to get my head around.
        ::Params
        inputs - numpy array of input features
        probs_actual - numpy array of actual probabilities of the training data
        e - the epsilon parameter to add or subtract from weights to calculate
        the slope of the loss function.
        ::Returns
        An array of gradients, one for each input and one extra for the bias
        '''
        
        # Preserve actual weights - we are going to change them for computation
        orig_weights = self.weights
        
        numgrad = np.zeros(self.weights.shape)
        perturb = np.zeros(self.weights.shape)

        for p in range(len(self.weights)):
            
            # Set perturbation vector
            perturb[p] = e
        
            self.weights = orig_weights + perturb
            probs_predicted = self.evaluate(inputs)
            loss2 = self.loss(probs_actual, probs_predicted)
            
            self.weights = orig_weights - perturb
            probs_predicted = self.evaluate(inputs)
            loss1 = self.loss(probs_actual, probs_predicted)

            # Compute numerical gradient
            numgrad[p] = (loss2 - loss1) / (2 * e)

            # We have to change the value for this weight index to zero
            perturb[p] = 0
       
        self.weights = orig_weights
    
        return numgrad

 
    def update_model(self, inputs: np.ndarray, probs_actual: np.ndarray,
                     lr: float, e_delta: float) -> np.ndarray:
        '''
        Implement an update function for the model that calculates gradients
        for our perceptron's weights/bias and returns a new set of weights/bias
        that reduces loss.
        ::Params
        inputs - numpy array of input features
        probs_actual - numpy array of actual probabilities of the training data
        lr - learning rate, a float value
        e_delta - epsilon parameter for caclulating gradients
        ::Returns
        A numpy array of weights, one for each input and one extra for the bias
        '''

        # Calculating gradients first
        grads = self.gradients(inputs, probs_actual, e_delta)

        # Applying gradients to our weights and bias
        self.weights = self.weights - lr * grads
        
        return self.weights


    def fit(self, inputs: np.ndarray, probs_actual: np.ndarray, lr: float,
            e_delta: float, iterations: int) -> tuple[float, float, list[float]]:
        '''
        Implements the fitting algorithm. At the moment the only stop-condition
        is the number of iterations. I will think about a loss-decrease-threshold
        for the next version.
        Please note: this function does not generate random weights and bias. It either
        takes the default values (zeroes) created in the constructor or weights can be set
        by using the set_params() function before fitting.
        ::Params
        inputs - numpy array of input features
        probs_actual - numpy array of actual probabilities of the training data
        lr - learning rate, a float value
        e_delta - epsilon parameter for calculating gradients
        iterations - how many iterations to perform
        ::Returns
        A numpy array of new weights and the history of losses
        '''

        losses = []

        # Iterating until the number of epochs
        for epoch in range(iterations):
            
            # predictions = self.predict(features, targets)
            self.update_model(inputs, probs_actual, lr, e_delta)
            
            # Predicted probabilities
            probs_pred = self.evaluate(inputs)
            
            # Current loss
            curr_loss = self.loss(probs_actual, probs_pred)
            losses.append(curr_loss)

        return self.weights, losses


    # Helper functions for interacting with outside world
    def get_params(self):
        '''
        Returns the weights vector - last element is the bias.
        '''
        # Get weights
        params = self.weights
        
        return params
    
    
    def set_params(self, weights, bias):
        '''
        Sets the weights vector and bias - usually before fitting.
        '''
        # Set weights+bias into one array
        self.weights = np.append(weights, bias)

        
```

```python
feateng_cosx = np.cos(feateng_x)
feateng_cosy = np.cos(feateng_y)
feateng_data = np.column_stack((feateng_x, feateng_y, feateng_cosx, feateng_cosy, feateng_group))
```

If we use these 4 features to try and $group$, the ANN will have a far easier time. Instead of trying to predict $group$ simply from weighted sums of $x$ and $y$, the ANN will try to predict $group$ from weighted sums of $x$, $cos(x)$, $y$ and $cos(y)$. And it should quickly find that the best fit only uses $cos(x)$ and $cos(y)$, and converge until it finds the combination of weights and bias that work best. While this is **not** part of thie E-tivity, feel free to experiment on the side and try to fit an ANN to the data above, first only with $x$ and $y$, then with the enriched data. The latter should perform far better.

Of course, when trying to fit some data, you will rarely know in advance what specific features will lead to a better fit. So there is an exercise in trying to think of what the shape of the data might come from, as well as some trial and error generating additional features and trying to fit models. But the common thread of engineered features is that the ones likely to provide considerable improvements are the features that can't trivially be calculated from a weighted sum. There would be no point to create a feature for $2\times x$ and $3\times x$ for example, as the ANN will naturally try various weights to multiply $x$ by.
